```python

# 변수와 자료형 - 문자열, 정수, 실수, 불리언
name = "한승현"
age = 58
height = 180
is_teacher = True

print(name)
print(age)
print(height)
print(is_teacher)
```

```python
# 변수 활용 - 계산과 f-string
next_age = age + 1
print(next_age)

greeting = "안녕하세요," + name + "선생님!"
print(greeting)

greeting2 = f"안녕하세요, {name}선생님! 나이는 {age}살이에요."
print(greeting2)
```

```python
# 리스트 - 여러 값을 한 상자에 담기
subjects = ["인공지능", "드론", "3D프린팅"]

# 리스트 전체 출력
print(subjects)

# 하나씩 꺼내기 (순서는 0부터 시작해요!)
print(subjects[0])
print(subjects[1])
print(subjects[2])
```

```python
# 리스트 활용 - 추가, 길이, 마지막 항목
subjects = ["인공지능","드론","3D프린팅"]

# 리스트에 항목 추가하기
subjects.append("유튜브")
print(subjects)

# 리스트 길이(항목이 몇 개인지)
print(len(subjects))

# 마지막 항목 꺼내기
print(subjects[-1])
```
